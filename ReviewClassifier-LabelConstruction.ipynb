```python
%%R -i df_English

corp = df_English$Review_lower

toks <- tokens(corp, remove_punct = TRUE, remove_symbols = TRUE,
               remove_numbers = TRUE, remove_url = TRUE)


tokens_replaced <- tokens_replace(toks, pattern = lexicon::hash_lemmas$token, replacement = lexicon::hash_lemmas$lemma)
## Lemmatized

dfmt <- dfm(tokens_replaced) |>
    dfm_remove(stopwords("en")) |> ## Stop words removed
    dfm_remove("*@*") |>
    dfm_trim(max_docfreq = 0.4, ## To avoid common words like "a" dominating. Arbitrary
             docfreq_type = "prop")
```

```python
keywords = [
'service',
'customer',
'ask',
'cashier',
'employee',
'staff',
'manager',
'friendly',
'self-checkout',
'experience',
'owner',
'follow',
'problem',
'request',
'easily',
'people',
'say'
]

#note: stuff like "nice" which can describe service seemed to be used more like "nice food"
```

```python
%%R -i df_English -i keywords

dfmt_df <- convert(dfmt, to = "data.frame")

is_service_df <- dfmt_df %>%
  tibble() %>%
  rowwise() %>%
  mutate(is_service = ifelse(sum(c_across(keywords) > 0) > 1, TRUE, FALSE)) %>%
    #If there are at least 2 service "keywords" as defined
    #You can edit this
  ungroup()%>%
  select(doc_id, is_service)
```

```python
%%R

is_service_df%>%
    count(is_service)

## This is a weak rule labelled by match frequency
```

```python
%%R -i df_English

my_df_English <- df_English%>%add_column(is_service_df$is_service)%>%rename(is_service = `is_service_df$is_service`)
str(my_df_English)
```
